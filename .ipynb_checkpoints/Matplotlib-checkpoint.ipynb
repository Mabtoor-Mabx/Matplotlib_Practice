#   Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6],[6,4,2])
plt.show()
```

```python
x = [2,4]
y = [5,5]
plt.plot(x,y)
plt.show()
```

```python
x = [2,4,6,8,10]
y = [1,3,5,7,9]
plt.plot(x,y, label='X Y label', linewidth=3)
plt.title('XY Plot Table', color='red')
plt.xlabel('X Plot')
plt.legend()
plt.ylabel('Y Plot')

plt.show()
```

```python
x= [1,2,3,4,5]
y=[6,7,8,9,10]
x1=[1,3,5,7,9]
y1=[2,4,6,8,10]
x2=[2,3,5,7,11]
y2=[1,4,7,10,13]

plt.plot(x,y, label='X and Y plots' ,color='Red', linewidth=4)
plt.plot(x1,y1,label='X1 and Y1 plots', color='Blue', linestyle='--')
plt.plot(x2,y2,label='X2 and Y2 plots', color='Orange', markersize=4 )
plt.title('X, X1 X2 & Y, Y1 Y2 Plots', fontdict = {'fontname': 'Times New Roman'})
plt.xlabel('X, X1 & X2 Plots')
plt.ylabel('Y, Y1 &  Y2 Plots')
plt.xticks((2,4,6,8,10,12,14,16))
plt.yticks((1,3,5,7,9,11,13,15))
plt.legend()

plt.show()

```

```python
plt.plot(x,y,'r^--', label='2x', markeredgecolor='blue')
```

```python
plt.figure(figsize=(6,8))

x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2, 'r', label='X^2')
plt.savefig('Info.png', dpi=100)
```
